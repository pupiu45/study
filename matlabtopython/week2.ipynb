#### 1. 산술 연산 : 기본과 적용들
#### 영상 처리에 관한 산술 연산들은 보통 픽셀 단위로 이루어진다. 그래서 연산은 각각의 픽셀들에 독립적으로 처리가 된다. 2D 배열 X와 2D 배열(혹은 스칼라) Y 그리고 결과 배열 Z는 연산 opn(+,-,*,/)에 대해
#### X opn Y=Z
#### 로 나타내질 수 있다.
#### - 덧셈 연산
#### 덧셈 연산은 두 영상 이미지를 섞거나 영상의 각 픽셀에 상수값을 더해주는데 사용된다. 0이 검정색이고 값이 커지면 커질수록 밝기가 밝아지므로, 덧셈 연산은 영상의 전체적 밝기에 영향을 끼친다. 또한 영상의 각 픽셀에 난수를 더해주는 것은 영상에 노이즈를 추가할 수 있는 대표적인 방법이다. 이런 노이즈를 없애는 방법은 Ch12에서 다룬다.
#### 영상을 더해주는 방법에는 두 가지가 있는데, 첫 번째는 정규화, 두 번째는 절단이다. 
#### 정규화는 영상의 최댓값과 최솟값을 픽셀의 자료형의 최댓값과 최솟값으로 맞춰주는 작업이다. 픽셀의 결과값을 g, 정규화 전의 값을 f, 전체 영상의 최댓값을 f_max, 최솟값을 f_min, 자료형의 최댓값을 L_max라 할 때, 픽셀의 정규화된 값은
$g=\frac{L_max}{(f_max-f_min)} (f-f_min)$
#### 로 나타낼 수 있다. $\frac{(f-f_min⁡)}{(f_max-f_min )}$의 값이 0 이상 1 미만이기 때문에, g는 0에서 L_max까지의 범위를 가진다.
#### 절단은 자료형의 최댓값보다 큰 값은 최댓값으로, 최솟값보다 작은 값은 최솟값으로 지정해주는 작업이다. MATLAB에서는 imadd라는 함수를 통해 쉽게 이 작업을 할 수 있다.
#### - 뺄셈 연산
#### 뺄셈 연산은 두 영상간의 차이를 찾아낼 때 이용된다. 이런 차이들은 원 이미지에의 인공적인 추가나 삭제, 비디오에서의 두 프레임간의 움직임 등등에 의해 발생한다.
#### 이미지에서 상수(스칼라)를 빼는 행위는 전체적 밝기의 감소를 야기한다. 이 때 생길 수 있는 음수의 문제를 다루는 것이 중요하다.
#### 이 underflow 문제를 해결할 수 있는 방법은 두 가지가 있다. 첫 번째는 절댓값을 취해 양수값만을 취하는 것이고, 두 번째는 역시 음수를 0으로 치환하는 절단이다.
#### MATLAB의 이미지 프로세싱 툴박스는 영상(행렬)간의 뺄셈을 할 수 있는 imsubtract 함수와 절댓값을 취하는 imabsdiff 함수를 제공한다. 이름에서 알 수 있듯 imabsdiff는 절댓값을, imsubtract는 절단을 통해 음수값을 다룬다. 
#### 또한 영상의 음수값을 취하는 imcomplement함수도 제공한다. 이것은 각 픽셀에 $g=-f+L_max$를 적용함으로써 얻을 수 있다.

#### - 곱연산과 나눗셈 연산
#### 영상에 스칼라를 곱하고 나눠주는 작업은 영상의 밝기를 조정하기 위해 자주 이용된다. 1보다 큰 수를 영상에 곱해주면 영상이 더 밝아지고, 0에서 1 사이의 값을 곱해주면 영상이 더 어두워진다. 이를 통해 가산적인 연산보다 더 영상의 밝기를 쉽게 다룰 수 있다.
#### MATLAB의 이미지 프로세싱 툴박스는 영상과 영상, 혹은 영상과 스칼라를 곱할 수 있는 함수인 immultiply를 제공한다. 또한 영상을 스칼라로 나눌 수 있는 imdivide도 제공한다.
#### - 간단한 산술 연산들
#### 영상들에 몇 가지 산술 연산들을 함께 사용하는 것을 배우는 것은 중요하다. 몇몇 연산들은 오버플로우나 언더플로우를 야기할 수 있기 때문이다. 둘 혹은 그 이상의 영상들의 일차결합을 수행할 때 절단과 반올림 등을 외연적으로 야기하는 것을 피하기 위해 이미지 프로세싱 툴박스는 imlincomb 함수를 제공한다. imlincom함수는 출력물을 독립적으로 계산하고, 만일 출력 배열이 정수형이라면 반올림을 수행해주고 범위를 초과하면 절단해준다.
#### 2. 논리 연산들 : 기본과 적용들
#### 논리 연산들은 각각의 픽셀값들의 이진 데이터들을 비트별로 다루기 위해 사용된다. AND, OR, XOR 등은 두개 이상의 인수가 필요한데 반해 NOT 연산은 하나의 인수만이 필요하다. 이진 영상에서는 보통 true(1)을 흰색, false(0)을 검은색으로 표현한다.
#### MATLAB에서는 행렬간의 논리연산을 위한 bitand, bitor, bitxor, bitcmp 함수들을 제공한다.

#### roipoly를 구현할 줄 몰라 따로 만든 이미지로 bw를 대체했습니다.
#### &을 통해 bitand()와 동일한 수행을 할 수 있다. 마스크가 TRUE인 부분만 영상이 표시되고, 나머지는 검은 색이 표시된다.
#### MATLAB과는 다르게, 255와 1 모두 true임을 알 수 있다.

```python
bw_cmp=~bw2
plt.subplot(1,2,1)
plt.imshow(bw,cmap=plt.cm.gray)
plt.subplot(1,2,2)
plt.imshow(bw_cmp,cmap=plt.cm.gray)
plt.show()
```

#### ~ 연산자를 통해 bitcmp()의 역할을 할 수 있다. bitcmp()는 영상을 뒤집기 위해 쓴다. TRUE(255)인 부분은 FALSE(0), FALSE는 TRUE가 된다.

```python
I3=I|bw_cmp
plt.imshow(I3,cmap=plt.cm.gray)
plt.show()
```

#### |연산자를 통해 bitor의 역할을 할 수 있다. 마스크가 true(255)인 부분은 흰색, FALSE(0)인 부분은 영상이 표현된다.

```python
I=plt.imread('cameraman.tif')
I2=plt.imread('cameraman2.tif')
I_xor=I^I2
plt.subplot(1,3,1)
plt.imshow(I,cmap=plt.cm.gray)
plt.subplot(1,3,2)
plt.imshow(I2,cmap=plt.cm.gray)
plt.subplot(1,3,3)
plt.imshow(I_xor,cmap=plt.cm.gray)
plt.show()
```

#### imabsdiff로 두 이미지의 차이를 구별할 수 있지만, ^(XOR)연산을 통해서도 두 이미지의 차이를 알 수 있다.
#### 이렇게 논리 연산들은 특정한 일을 수행하기 위해 쓰인다. 다음은 한 부분만 어둡게 하기 위해 논리 연산을 사용한다.

```python
I=plt.imread('lindsay.tif')
I_adj=(I/1.5).astype('uint8')
bw=plt.imread('lindsay_bw.bmp')
r,g,b=bw[:,:,0],bw[:,:,1],bw[:,:,2]
bw2= ((0.2989 * r + 0.5870 * g + 0.1140 * b)*255).astype('uint8')
bw_cmp=~bw2
original=I_adj|bw_cmp
dark=I|bw2
new_img=original&dark
plt.imshow(new_img,cmap=plt.cm.gray)
plt.show()
```
